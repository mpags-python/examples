```python
import numpy as np
```

```python
x = np.array([ 1,  3,  9, 14, 15, 19, 20, 21, 30, 35,
              40, 41, 42, 47, 54, 56, 67, 69, 72, 88])

y = np.array([33, 68, 34, 34, 37, 65, 37, 44, 48, 49,
              53, 49, 50, 48, 56, 60, 61, 63, 44, 71])

e = np.array([3.6, 3.9, 2.6, 3.4, 3.8, 14.8, 2.2, 2.1, 2.3, 3.8,
              2.2, 2.8, 3.9, 3.1, 3.4, 2.6, 3.4, 3.7, 2.0, 3.5])
```

### matplotlib notebook

```python
%matplotlib notebook
import matplotlib.pyplot as plt
fig = plt.figure()
plt.scatter(x, y, 1000/e**2)
plt.show()
```

```python
plt.xlabel('resistance, $\Omega$ ohms')
plt.ylabel('size of universe, $S$ Gpc')
plt.tight_layout()
```

### mpld3

```python
import mpld3
mpld3.display(fig)
```

```python
mpld3.save_html(fig, open('scatter_mpld3.html', 'w'))
```

### bokeh

```python
import bokeh.plotting as bk
bk.output_notebook()
p = bk.figure()
p.scatter(x,y, radius=5/e, line_color=None)
bk.show(p)
```

```python
bk.save(p, 'scatter_bokeh.html')
```
